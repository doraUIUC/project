```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_losses_and_errors(l1_train_losses, validation_errors):

    epochs = np.arange(1, len(l1_train_losses)+1)

    # Plot the L1 training losses
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, l1_train_losses, 'b', label='L1 Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('L1 Training Loss over Epochs')
    plt.legend()
    plt.grid()
    plt.show()

    # Plot the 3px validation errors
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, validation_errors, 'r', label='3px Validation Error')
    plt.xlabel('Epochs')
    plt.ylabel('Error')
    plt.title('3px Validation Error over Epochs')
    plt.legend()
    plt.grid()
    plt.show()


```

```python
l1_finetune = np.load('results/finetune_train_loss_epoch.npy')
val_finetune = np.load('results/finetune_val_err_epoch.npy')
```

```python
visualize_losses_and_errors(l1_finetune, val_finetune)
```

```python
l1_resnet = np.load('results/resnet_finetune_train_loss_epoch.npy')
val_resnet = np.load('results/resnet_finetune_val_err_epoch.npy')
```

```python
visualize_losses_and_errors(l1_resnet, val_resnet)
```
